```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
```

```python
tl=pd.read_csv('custom_commands2.csv')
```

```python
tl
```

```python
x=tl['Custom Command']
y=tl['Windows CMD Equivalent']
```

```python
x
```

```python
y
```

```python
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier()
```

```python
le_x = LabelEncoder()  # Encoder for custom commands
le_y = LabelEncoder()  # Encoder for Windows commands

x_encoded = le_x.fit_transform(x)
y_encoded = le_y.fit_transform(y)
```

```python
x_encoded
```

```python
y_encoded
```

```python
x_encoded = np.array(x_encoded).reshape(-1, 1)
y_encoded = np.array(y_encoded)  # No need to reshape y
```

```python
model.fit(x_encoded, y_encoded)  # Train the model
```

```python
predicted = model.predict([[0]])  # Predict for encoded value 0
output = le_y.inverse_transform(predicted)[0]  # Use le_y, NOT le_x
print(output)
```

```python
def answer(command):
    encoded_cmd = le_x.transform([command])  # Encode using le_x
    prediction = model.predict(np.array(encoded_cmd).reshape(-1, 1))  # Ensure correct shape
    return le_y.inverse_transform(prediction)[0]  # Decode using le_y
```

```python
pred=answer('jao')
```
